### Importación de datos



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()

```

#1. Análisis de facturación



```python
#Total de ventas tienda 1
tienda.isna().sum()
ingresosTienda1 = tienda['Precio'].sum()
ingresosTienda1
formatodinero = "${:,.2f}".format(ingresosTienda1)
formatodinero
```

```python
#Total ventas tienda 2
ingresosTienda2 = tienda2['Precio'].sum()
ingresosTienda2
formatodinero2 = "${:,.2f}".format(ingresosTienda2)
formatodinero2

```

```python
#Total ventas tienda 3
ingresosTienda3 = tienda3['Precio'].sum()
ingresosTienda3
formatodinero3 = "${:,.2f}".format(ingresosTienda3)
formatodinero3
```

# 3. Calificación promedio de la tienda


```python
#Calificacion  promedio tienda 1
calificacion = tienda.groupby('Producto')['Calificación'].mean().round(2)
calificacion.mean().round(2)
```

```python
#Calificacion por producto tienda 1
calificacion.sort_values(ascending=False)

```

```python
#Calificacion promedio tienda 3
calificacion3 = tienda3.groupby('Producto')['Calificación'].mean().round(2)
calificacion3.mean().round(2)
```

```python
#Calificacion por producto tienda 3
calificacion3.sort_values(ascending=False)
```

```python
#Calificacion promedio tienda 4
calificacion4 = tienda4.groupby('Producto')['Calificación'].mean().round(2)
calificacion4.mean().round(2)
```

# 4. Productos más y menos vendidos

```python
dftienda1 = tienda[['Producto', 'Precio']].copy()
total_ventas1 = dftienda1.groupby('Producto')['Precio'].sum()
pmasvendidotienda1_sorted = total_ventas1.sort_values(ascending=False)
pmenosvendidotienda1_sorted = total_ventas1.sort_values(ascending=True)
print(f'El producto más vendido de la tienda 1 fue: {pmasvendidotienda1_sorted.index[0]} con un valor total de ventas de: ${pmasvendidotienda1_sorted.iloc[0].astype(int):,}')
print(f'El producto menos vendido de la tienda 1 fue: {pmenosvendidotienda1_sorted.index[0]} con un valor total de ventas de: ${pmenosvendidotienda1_sorted.iloc[0].astype(int):,}')
pmasvendidotienda1_sorted.plot(kind='bar', figsize=(12,6), title='Total ventas por producto', xlabel='Producto', ylabel='Precio')
```

```python
dftienda2 = tienda2[['Producto', 'Precio']].copy()
total_ventas2 = dftienda2.groupby('Producto')['Precio'].sum()
pmasvendidotienda2_sorted = total_ventas2.sort_values(ascending=False)
pmenosvendidotienda2_sorted = total_ventas2.sort_values(ascending=True)
print(f'El producto más vendido de la tienda 2 fue: {pmasvendidotienda2_sorted.index[0]} con un valor total de ventas de: ${pmasvendidotienda2_sorted.iloc[0].astype(int):,}')
print(f'El producto menos vendido de la tienda 2 fue: {pmenosvendidotienda2_sorted.index[0]} con un valor total de ventas de: ${pmenosvendidotienda2_sorted.iloc[0].astype(int):,}')
pmasvendidotienda2_sorted.plot(kind='bar', figsize=(12,6), title='Total ventas por producto', xlabel='Producto', ylabel='Precio')
```
